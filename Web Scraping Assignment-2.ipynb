## Q1

```python
# opening webpage 'https://www.naukri.com/'

driver.get('https://www.naukri.com/')
```

```python
#enter Job role 'Data Analyst' in input box

jobsearch = driver.find_element(By.CLASS_NAME,"suggestor-input ")
jobsearch.send_keys('Data Analyst')
```

```python
#enter Location 'Bangalore' in location input box

location = driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location.send_keys('Bangalore')
```

```python
enter = driver.find_element(By.CLASS_NAME,'qsbSubmit')
enter.click()
```

```python
job_title = []
job_location = []
company_name = []
experience_required = []
```

```python
# scraping job title 
title_tag = driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')

for i in title_tag[0:10]:
    title = i.text
    job_title.append(title)
    
#scraping Location
location_tag = driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')

for i in location_tag[:10]:
    locat = i.text
    job_location.append(locat)
    
# Scraping campany Name
comName_tag = driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')

for i in comName_tag[:10]:
    name = i.text
    company_name.append(name)
    
# Scraping Experience
exp_tag = driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')

for i in exp_tag[:10]:
    experience = i.text
    experience_required.append(experience)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
df = pd.DataFrame({'Title':job_title,'Location':job_location,'Company Name':company_name,'Experience Required':experience_required})
```

```python
# Closing OTP dialog box

closeDialog = driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _2doB4z"]')
closeDialog.click()
```

```python
# enter our search product 'sunglasses'

enterPro = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
enterPro.send_keys('sunglasses')
```

```python
# click search icon
clicksearchicon = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
clicksearchicon.click()
```

```python
# create empty list of required data
brand = []
productDecrip = []
price = []
```

```python
# Scraping 100 sunglasses brand name, product description,and price

for page in range(0,3):
    brandName = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]') # extract brand Name
    for i in brandName:
        brand.append(i.text)
    pro_des = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]') # extract Product description
    for i in pro_des:
        productDecrip.append(i.text)
    pri = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]') # extract price
    for i in pri:
        price.append(i.text)
    next_button = driver.find_element(By.XPATH,'//a[@class="ge-49M"]') #click next page
    next_button.click()
    time.sleep(3)
```

```python
# creating Data Frame
GlassesData = pd.DataFrame({'Brand Name':brand[:100],'Product Description':productDecrip[:100],'Price':price[:100]})
GlassesData
```

## Q8

```python
# opening webpage with url 'https://www.azquotes.com/'
driver.get('https://www.azquotes.com/')
```

```python
# clicking Top Quotes
clickTop = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
clickTop.click()
```

```python
# creating List
quotes = []
authors = []
typeOfQuotes = []
```

```python
# scrap quates, authores and type of quote
for page in range(0,10):
    quote_all = driver.find_elements(By.XPATH,'//a[@class="title"]') #scrap Quotes
    for i in quote_all:
        quotes.append(i.text)
    
    authors_all = driver.find_elements(By.XPATH,'//div[@class="author"]') # Scrap Authors
    for i in authors_all:
        authors.append(i.text)
    
    typeQuote = driver.find_elements(By.XPATH,'//div[@class="tags"]') # Scrap type of quotes
    for i in typeQuote:
        typeOfQuotes.append(i.text)
    next_button = driver.find_element(By.CLASS_NAME,'next')
    try:
        next_button.click()
    except:
        pass
    time.sleep(3)
    
```

```python
# Creating Data Frame
All_Quotes = pd.DataFrame({'Quote':quotes,'Author':authors,'Type of Quote':typeOfQuotes})
All_Quotes
```

```python
# clicking GK option
GKoption = driver.find_element(By.XPATH,'/html/body/div/div[1]/div/div[1]/div/div[6]/div/div[1]/header/div[3]/ul/li[9]/a')
GKoption.click()
```

```python
# Clicking on List of all prime minister of India
all_pm = driver.find_element(By.XPATH,'/html/body/div/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a')
all_pm.click()
```

```python
# Create list of all required data
pm_name = []
pm_borndead = []
pm_office = []
pm_remarks = []
```

```python
# scrap PM name
pm_names = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[2]') # Prime Minister Manes
for i in pm_names:
    pm_name.append(i.text)
pm_bd = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[3]') # scraping Borth and death 
for i in pm_bd:
    pm_borndead.append(i.text)
pm_office_li = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[4]') # Scraping term of office
for i in pm_office_li:
    pm_office.append(i.text.replace('\n',' '))
pm_rem = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[5]') # scraping Remarks
for i in pm_rem:
    pm_remarks.append(i.text)
```

```python
print(len(pm_name),len(pm_borndead),len(pm_office),len(pm_remarks))
```

```python
Prime_Minister_of_India = pd.DataFrame({'Name':pm_name[:18],'Born-Dead':pm_borndead,'Term of Office':pm_office,'Remarks':pm_remarks})
Prime_Minister_of_India
```
