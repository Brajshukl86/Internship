```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests

import warnings
warnings.filterwarnings('ignore')
```

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

soup = bs(url.content)
```

```python
tbody = []
tables = soup.find_all('table')
for table in tables:
    tbody.append(table.find('tbody'))
```

```python
youtub = []
for td in tbody[1]('td'):
    youtub.append(td.text.replace('\n','').strip())
```

```python
ranks = []
video_name = []
uploader = []
date = []
views = []
```

```python
for i in range(0,len(youtub),6):
    ranks.append(youtub[i])
    video_name.append(youtub[i+1].split('[')[0])
    uploader.append(youtub[i+2])
    date.append(youtub[i+4])
    views.append(youtub[i+3])
```

```python
print(len(ranks),len(video_name),len(uploader),len(date),len(views))
```

```python
most_view_youtube_video = pd.DataFrame({'Rank':ranks,'Name':video_name,'Artist':uploader,'Upload date':date,'Views':views})
most_view_youtube_video
```

```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from bs4 import BeautifulSoup as bs
import requests
import time

import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r'chromedriver.exe')
```

```python
driver.maximize_window()
```

```python
driver.get('https://www.bcci.tv/')
```

```python
driver.find_element(By.XPATH,'//div[@class="collapse navbar-collapse"]/ul/li[2]').click()
```

```python
driver.find_element(By.XPATH,'//div[@class="cookie no-disply"]/button').click()
```

```python
driver.find_element(By.XPATH,'//div[@class="col-lg-12 col-md-12 col-sm-12"]/button').click()
```

```python
table = driver.find_element(By.CLASS_NAME,'fixture-tab-inner').text.split('\n')
```

```python
match_data = []

while len(table) == len(table):
    try:
        
        if table[2] == 'LIVE':
            live = []
            for i in range(0,11):
                live.append(table.pop(0))
            match_data.append(live)
        elif table[2] != 'LIVE':

            upcoming = []
            for i in range(0,8):
                upcoming.append(table.pop(0))
            match_data.append(upcoming)
    except:
        break
```

```python
match_title = []
series = []
place = []
date = []
time = []

for i in match_data:
    series.append(i[0])
    date.append(i[1])
    time.append(i[2])
    place.append(i[-2].split('-')[1].strip())
    match_title.append(i[-2].split('-')[0].strip())
```

```python
print(len(match_title),len(series),len(place),len(date),len(time))
```

```python
series_data = pd.DataFrame({'Match Title':match_title,'Series':series,'Place':place,'Date':date,'Time':time})
series_data
```

```python
driver.close()
```

```python
imdb_data = imdb_soup.find_all('div',{'class':'lister-item-content'})
```

```python
imdb_data = driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]')
```

```python
imdb_name = []
imdb_year = []
imdb_genre = []
imdb_run = []
imdb_ratings = []
imdb_votes = []
for i in imdb_data:
    imdb_name.append((i.find_element(By.CLASS_NAME,'lister-item-header').text).split('.')[1].split('(')[0].strip())
    imdb_year.append((i.find_element(By.CLASS_NAME,'lister-item-header').text).split('(')[1].split(')')[0].strip())
    imdb_genre.append(i.find_element(By.CLASS_NAME,'genre').text.strip())
    imdb_run.append(i.find_element(By.CLASS_NAME,'runtime').text.strip())
    imdb_ratings.append(i.find_element(By.CLASS_NAME,'ipl-rating-star__rating').text.strip())
    imdb_votes.append(i.find_element(By.NAME,'nv').text.strip())
```

```python
print(len(imdb_name),len(imdb_year),len(imdb_genre),len(imdb_run),len(imdb_ratings),len(imdb_votes))
```
