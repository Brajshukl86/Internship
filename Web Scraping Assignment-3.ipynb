```python
# import all neccesories libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import io
from PIL import Image
import time
```

```python
# open Chrome
driver = webdriver.Chrome(r'chromedriver.exe')
```

```python
# chrome window maximize
driver.maximize_window()
```

```python
# Get user input
google_search = input('Enter Keyword: ')
```

```python
# Make fuction for image and download
def get_images_from_google(driver, max_images):
    def scroll_down(driver):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)
        
    driver.get('https://images.google.com/')

    enter_keyword = driver.find_element(By.XPATH,'//input[@class="gLFyf"]')
    enter_keyword.send_keys(google_search)
    time.sleep(5)
    
    search_button = driver.find_element(By.XPATH,'//button[@class="Tg7LZd"]/div')
    search_button.click()
    time.sleep(10)

    images_urls = set()
    skips = 0

    while len(images_urls) + skips < max_images:
        scroll_down(driver)

        thumbnail = driver.find_elements(By.CLASS_NAME,'Q4LuWd')

        for img in thumbnail[len(images_urls) + skips:max_images]:
            try:
                img.click()
                time.sleep(5)
            except:
                continue

            images = driver.find_elements(By.CLASS_NAME,'n3VNCb')
            for image in images:

                if image.get_attribute('src') and 'http' in image.get_attribute('src'):
                    images_urls.add(image.get_attribute('src'))


    return images_urls

def download_image(download_path, url, file_name):            #create function for download image
    try:
        image_content = requests.get(url).content
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file)
        file_path = download_path + file_name
        
        with open(file_path, "wb") as f:
            image.save(f, "jpeg")
            
        
    except Exception as e:
        print('Failed -', e)
        
        
urls = get_images_from_google(driver, 10) # call fuction

for i, url in enumerate(urls):
    download_image("images/", url, str(i)+".jpg")
```

```python
# close Chrome 
driver.close()
```

```python
# Make list for all required data
brand_name = []
Model_name =[]
colour = []
RAM = []
strorage = []
primary_camera = []
secondry_camera = []
display_size = []
battery_capacity = []
price = []
```

```python
# Fetch all data with beautifulsoup
items = []
x_data = []
for url in product_url:
    r = requests.get(url)
    soup = bs(r.content)
    brand_name.append(soup.find('h1',{'class':'yhB1nd'}).text.replace('\xa0',''))
    price.append(soup.find('div',{'class':'_30jeq3 _16Jk6d'}).text)
    table = soup.find_all('tr',class_="_1s_Smc")
    for td in table:
        x_data.append(td.find('td',{'class':'_1hKmbr col col-3-12'}).text)
        items.append(td.find('li').text)
    
    try:
        if 'Model Name' in x_data:
            Model_name.append(items[x_data.index('Model Name')])
        else:
            Model_name.append(np.nan)
    except Exception as e:
        Model_name.append(np.nan)
    
    try:
        if 'Color' in x_data:
            colour.append(items[x_data.index('Color')])
        else:
           colour.append(np.nan) 
    except Exception as e:
        colour.append(np.nan)
    
    try:
        if 'RAM' in x_data:
            RAM.append(items[x_data.index('RAM')])
        else:
            RAM.append(np.nan)
    except Exception as e:
        RAM.append(np.nan)
    
    try:
        if 'Internal Storage' in x_data:
            strorage.append(items[x_data.index('Internal Storage')])
        else:
            strorage.append(np.nan)
    except Exception as e:
        strorage.append(np.nan)
    
    try:
        if 'Primary Camera' in x_data:
            primary_camera.append(items[x_data.index('Primary Camera')])
        else:
            primary_camera.append(np.nan)
    except Exception as e:
        primary_camera.append(np.nan)
    
    try:
        if 'Secondary Camera' in x_data:
            secondry_camera.append(items[x_data.index('Secondary Camera')])
        else:
            secondry_camera.append(np.nan)
    except:
            secondry_camera.append(np.nan)
    
    try:
        if 'Display Size' in x_data:
            display_size.append(items[x_data.index('Display Size')])
        else:
            display_size.append(np.nan)
    except Exception as e:
        display_size.append(np.nan)
    
    try:
        if 'Battery Capacity' in x_data:
            battery_capacity.append(items[x_data.index('Battery Capacity')])
        else:
            battery_capacity.append(np.nan)
    except Exception as e:
        battery_capacity.append(np.nan)
            

    
```

```python
driver = webdriver.Chrome(r'chromedriver.exe')
```

```python
driver.get('https://www.hostelworld.com/')
```

```python
hotel_search = driver.find_element(By.XPATH,'//div[@class="field-inner"]/input')
```

```python
hotel_search.send_keys('London')
```

```python
driver.find_element(By.XPATH,'//div[@class="label"]').click()
```

```python
hotel_click = driver.find_element(By.XPATH,'//div[@class="search-form layout-1"]/div/button')
hotel_click.click()
```

```python
hot_url = []
hotel_url = driver.find_elements(By.XPATH,'//div[@class="title-row"]/h2/a')
for i in hotel_url:
    hot_url.append(i.get_attribute('href'))
len(hot_url)
```

```python
hot_name = []
hot_rating = []
totalreviews = []
overallreviews = []
privatesprice = []
dormsprice = []
facilities_hot = []
propertydescription = []
for url in hot_url:
    hotel_u = requests.get(url)
    hote_soup = bs(hotel_u.content)
    hot_name.append(hote_soup.find('div',{'class':'title-2'}).text)
    try:
        hot_rating.append(hote_soup.find('div',{'class':'score orange big'}).text.replace('\n',''))
    except:
        hot_rating.append(np.nan)
    try:
        totalreviews.append(hote_soup.find('div',{'class':'score orange big'}).text.replace('\n',''))
    except:
        totalreviews.append(np.nan)
    try:
        overallreviews.append(hote_soup.find('div',{'class':'keyword'}).text)
    except:
        overallreviews.append(np.nan)
    try:
        privatesprice.append(hote_soup.find('div',{'id':'rate-199686-price'}).text)
    except:
        privatesprice.append(np.nan)
    try:
        dormsprice.append(hote_soup.find('div',{'class':'price'}).text)
    except:
        dormsprice.append(np.nan)
    facilities_hot.append(hote_soup.find('ul',{'class':'facility-sections'}).text.replace('\n',''))
    propertydescription.append(hote_soup.find('ul',{'class':'facility-sections'}).text.replace('\n',''))
```

```python
print(len(hot_name),len(hot_rating),len(totalreviews),len(overallreviews),len(privatesprice),len(dormsprice),len(facilities_hot),len(propertydescription))
```

```python
distance = []
hot_disc = driver.find_elements(By.XPATH,'//div[@class="subtitle body-3"]/a/span[1]')
for i in hot_disc:
    distance.append(i.text.strip())
len(distance)
```

```python
hotel_df = pd.DataFrame({'Hostel Name':hot_name,'Distance from City Centre':distance,'Ratings':hot_rating,'Total Reviews':totalreviews,'Overall Reviews':overallreviews,'Privates from Price':privatesprice,'Dorms from Price':dormsprice,'Facilities':facilities_hot,'Property Description':propertydescription})
hotel_df
```
