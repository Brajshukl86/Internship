# Q2- Ans

## Make Data Frame for 100 Top rating Movies

#### Load HTML code for all two pages

```python
imdb_page1 = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
imdb_page2 = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")
soup1 = bs(imdb_page1.content)
soup2 = bs(imdb_page2.content)

content_imdb = soup1.prettify()
content_imdb1 = soup2.prettify()

print(content_imdb)
```

#### Find Name and year Tag and store name and year as a list

```python
name1 = soup1.find_all("h3", class_="lister-item-header")
name2 = soup2.find_all("h3",class_="lister-item-header")
names = name1+name2
movie_name = []
movie_rea_date = []

for name in names:
    mov_name = (name.find("a")).text
    mov_year = ((name.find("span",class_="lister-item-year text-muted unbold")).text)
    
    movie_name.append(mov_name)
    movie_rea_date.append(mov_year)
    
print(movie_name)
print(movie_rea_date)

```

#### Find Rating in text and store as list

```python
ratings1 = soup1.find_all("div",class_="inline-block ratings-imdb-rating")
ratings2 = soup2.find_all("div",class_="inline-block ratings-imdb-rating")
ratings = ratings1+ratings2


movie_rats = []

for rate in ratings:
    rating = (rate.text).replace('\n','')
    movie_rats.append(rating)
print(movie_rats)
```

#### Make a dictionary with name, rating and year

```python
IMDB_top_100_movies1 = {'Name':movie_name,'Ratings':movie_rats,'Release Date':movie_rea_date}

IMDB_top_100_movies1
```

#### All out put store as a DataFrame

```python
IMDB_top_100_movies = pd.DataFrame.from_dict(IMDB_top_100_movies1)
```

```python
IMDB_top_100_movies
```

# Q3- Ans

### Load HTML code of website

```python
url = requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
soup = bs(url.content)

contents1 = soup.prettify()

print(contents1)
```

### extract Name and Year

```python
names = soup.find_all("td",class_="titleColumn")

mov_name = []
mov_year = []

for name in names:
    movi_nam = name.find('a').text
    movi_ye = name.find('span').text.split('(')[1].split(')')[0]
    mov_name.append(movi_nam)
    mov_year.append(movi_ye)
print(mov_name)
print(mov_year)
```

### Store Rating as list

```python
rating_in = soup.find_all("td",class_="ratingColumn imdbRating")

rating_mov = []

for rate in rating_in:
    ratin = rate.text.split('\n')[1]
    rating_mov.append(ratin)
print(rating_mov)
```

```python
IMDB_indian_movie_dic = {'Movie Name':mov_name[0:100],'Movie Rating':rating_mov[0:100],'Release Year':mov_year[0:100]}
```

```python
IMDB_indian_movie = pd.DataFrame.from_dict(IMDB_indian_movie_dic)
IMDB_indian_movie
```

```python
president = requests.get("https://presidentofindia.nic.in/former-presidents.htm")

soup = bs(president.content)

content_president = soup.prettify()
print(content_president)
```

```python
president_name = soup.find_all("div",class_="presidentListing")

print(president_name)
```

```python
former_president_name = []
term_of_office = []

for name in president_name:
    pre_name = name.find("h3").text.split('(',)[0]
    term_offic = name.find('p').text.split(':')[1]
    former_president_name.append(pre_name)
    term_of_office.append(term_offic)
print(former_president_name)
print(term_of_office)
```

# Q7

Fetch HTML code

```python
cbn_url = requests.get('https://www.cnbc.com/world/?region=world')

cbn_soup = bs(cbn_url.content)
```

Fetch target HTML content and filter Headline and Time text and News link URL

```python
head_cbn = cbn_soup.find_all('div',class_="LatestNews-headlineWrapper")

headline = []
time = []
news_link = []

for head in head_cbn:
    head_line = head.find('a',class_='LatestNews-headline').text.strip()
    post_time = head.find('time',class_='LatestNews-timestamp').text.strip()
    newsLink = head.find('a',class_='LatestNews-headline').get('href')
    headline.append(head_line)
    time.append(post_time)
    news_link.append(newsLink)
```

Here asign Dictionary

```python
News = {'Headline':headline,'Time':time,'News Link':news_link}
News
```

```python
art_url = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')

art_soup =bs(art_url.content)
```

```python
art_cont = art_soup.find_all('article')

paper_title = []
authors = []
publish_date = []
paper_url = []

for paper in art_cont:
    paperTitle = paper.find('h2',class_='sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg').text.strip()
    authors_paper = paper.find('span', class_='sc-1w3fpd7-0 dnCnAO').text.strip()
    publishDate = paper.find('span', class_='sc-1thf9ly-2 dvggWt').text.strip()
    paperURL = paper.find('a',class_='sc-5smygv-0 fIXTHm').get('href')
    paper_title.append(paperTitle)
    authors.append(authors_paper)
    publish_date.append(publishDate)
    paper_url.append(paperURL)
    
```

```python
Most_Download_article1 = {'Paper Title':paper_title,'Authors':authors,'Published Date':publish_date,'Paper URL':paper_url}

Most_Download_article = pd.DataFrame.from_dict(Most_Download_article1)
Most_Download_article
```

# Q9

```python
Restau_name = []
cuisine_name = []
dinLocation = []
dinRatings = []
dinImage = []
```

```python
for i in range(1,30):
    url = requests.get('https://www.dineout.co.in/delhi-restaurants?p='+str(i))

    din_soup = bs(url.content)

    din_text = din_soup.find_all('div',class_='restnt-main-wrap clearfix')

    for resto in din_text:
        restro_name = resto.find('a',class_='restnt-name ellipsis').text.strip()
        cuisine_food = resto.find('span',class_='double-line-ellipsis').text.split('|')[1]
        locat_ion = resto.find('div',class_='restnt-loc ellipsis').text
        if resto.find('div',class_='restnt-rating rating-4') == None:
            rat_ings = resto.find('div',class_='restnt-rating rating-4')
        else:
            rat_ings = resto.find('div',class_='restnt-rating rating-4').text
        imageURL = resto.find('img')['data-src']
        Restau_name.append(restro_name)
        cuisine_name.append(cuisine_food)
        dinLocation.append(locat_ion)
        dinRatings.append(rat_ings)
        dinImage.append(imageURL)
```

```python
Restraurant = pd.DataFrame({'Restaurant':Restau_name,'Cuisine':cuisine_name,'Location':dinLocation,'Rating': dinRatings,'Image URL':dinImage})
Restraurant
```
